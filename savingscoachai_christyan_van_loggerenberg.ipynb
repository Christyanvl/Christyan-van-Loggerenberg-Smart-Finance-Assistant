```python
# ============================================================
# [7] VALIDATION AND MONEY FORMAT FUNCTIONS
# ============================================================

def is_valid_text(text):
    text = str(text).strip()

    if text == "":
        return False

    return text.replace(" ", "").isalpha()


def format_money(amount):
    return f"Rs {amount:,.2f}"
```

```python
# ============================================================
# [8] MAIN SAVINGS GROWTH PLANNER FUNCTION
# ============================================================

def savings_growth_planner(
    saving_goal_name,
    saving_goal_amount,
    years_to_goal,
    monthly_goal_saving,
    monthly_income,
    groceries,
    transport,
    entertainment,
    rent_or_bills,
    other_expenses,
    bank_interest_rate,
    result_view
):

    if not is_valid_text(saving_goal_name):
        return "Error: Goal name must only contain letters.", pd.DataFrame(), "Error", "Error"

    saving_goal_amount = float(saving_goal_amount)
    years_to_goal = float(years_to_goal)
    monthly_goal_saving = float(monthly_goal_saving)
    monthly_income = float(monthly_income)
    groceries = float(groceries)
    transport = float(transport)
    entertainment = float(entertainment)
    rent_or_bills = float(rent_or_bills)
    other_expenses = float(other_expenses)
    bank_interest_rate = float(bank_interest_rate)

    if saving_goal_amount <= 0 or years_to_goal <= 0 or monthly_goal_saving <= 0:
        return "Error: Goal, years, and monthly saving must be more than 0.", pd.DataFrame(), "Error", "Error"

    if monthly_income <= 0:
        return "Error: Monthly income must be more than 0.", pd.DataFrame(), "Error", "Error"

    if bank_interest_rate < 0:
        return "Error: Interest rate cannot be negative.", pd.DataFrame(), "Error", "Error"

    total_monthly_expenses = (
        groceries
        + transport
        + entertainment
        + rent_or_bills
        + other_expenses
    )

    available_monthly_savings = (
        monthly_income - total_monthly_expenses
    )

    if available_monthly_savings < 0:
        return (
            "Your expenses are higher than your income.",
            pd.DataFrame(),
            "You cannot save yet.",
            "Reduce expenses first."
        )

    if monthly_goal_saving > available_monthly_savings:
        return (
            f"You chose to save {format_money(monthly_goal_saving)}, but only {format_money(available_monthly_savings)} is available.",
            pd.DataFrame(),
            "Chosen saving is too high.",
            "Lower monthly saving or reduce expenses."
        )

    total_months = int(years_to_goal * 12)

    monthly_interest_rate = (
        bank_interest_rate / 100
    ) / 12

    required_monthly_saving = (
        saving_goal_amount / total_months
    )

    running_total = 0

    table_data = []

    months_to_reach_goal = None

    yearly_saved = 0
    yearly_interest = 0

    for month in range(1, total_months + 1):

        running_total = (
            running_total + monthly_goal_saving
        )

        interest_income = (
            running_total * monthly_interest_rate
        )

        running_total = (
            running_total + interest_income
        )

        yearly_saved = (
            yearly_saved + monthly_goal_saving
        )

        yearly_interest = (
            yearly_interest + interest_income
        )

        progress = (
            running_total / saving_goal_amount
        ) * 100

        if progress > 100:
            progress = 100

        if running_total >= saving_goal_amount and months_to_reach_goal is None:
            months_to_reach_goal = month

        if result_view == "Monthly":

            table_data.append({
                "Period": f"Month {month}",
                "Amount Saved": format_money(monthly_goal_saving),
                "Interest Income": format_money(interest_income),
                "Total Savings": format_money(running_total),
                "Progress To Goal": f"{progress:.2f}%"
            })

        if result_view == "Yearly" and month % 12 == 0:

            year_number = month // 12

            table_data.append({
                "Period": f"Year {year_number}",
                "Amount Saved": format_money(yearly_saved),
                "Interest Income": format_money(yearly_interest),
                "Total Savings": format_money(running_total),
                "Progress To Goal": f"{progress:.2f}%"
            })

            yearly_saved = 0
            yearly_interest = 0

    savings_table = pd.DataFrame(table_data)

    expenses = {
        "Groceries": groceries,
        "Transport": transport,
        "Entertainment": entertainment,
        "Rent or Bills": rent_or_bills,
        "Other Expenses": other_expenses
    }

    highest_category = max(
        expenses,
        key=expenses.get
    )

    highest_amount = expenses[highest_category]

    possible_saving = (
        highest_amount * 0.20
    )

    if months_to_reach_goal is None:

        goal_time_text = f"""
According to my calculations, you will NOT reach your goal within {years_to_goal:.1f} year(s).

At your chosen saving amount of {format_money(monthly_goal_saving)} per month, you will have {format_money(running_total)} by the end of the plan.
"""

    else:

        goal_time_text = f"""
According to my calculations, you will reach your goal in about {months_to_reach_goal} month(s).

That is about {months_to_reach_goal / 12:.1f} year(s).
"""

    savings_advice = f"""
According to my calculations, your monthly income is {format_money(monthly_income)}.

According to my calculations, your total monthly expenses are {format_money(total_monthly_expenses)}.

According to my calculations, your available monthly savings are {format_money(available_monthly_savings)}.

According to my calculations, you chose to save {format_money(monthly_goal_saving)} per month.

According to my calculations, your highest expense category is {highest_category}, where you spend {format_money(highest_amount)} per month.

According to my calculations, if you reduce {highest_category} by 20%, you could save about {format_money(possible_saving)} more per month.
"""

    final_result = f"""
FINAL SAVINGS PLAN

Saving Goal:
{saving_goal_name}

Goal Amount:
{format_money(saving_goal_amount)}

Years Selected:
{years_to_goal:.1f}

Required Monthly Saving Without Interest:
{format_money(required_monthly_saving)}

Your Chosen Monthly Saving:
{format_money(monthly_goal_saving)}

Estimated Final Savings:
{format_money(running_total)}

Bank Interest Rate:
{bank_interest_rate:.2f}%

Result View:
{result_view}

Overall Recommendation:
According to my calculations, you should keep saving consistently and reduce your highest expense category to reach your goal faster.
"""

    return (
        savings_advice,
        savings_table,
        goal_time_text,
        final_result
    )
```

```python
# ============================================================
# [11] GRADIO APP
# ============================================================

with gr.Blocks() as app:

    gr.Markdown("# Savings Growth Planner")

    gr.Markdown("""
This app helps users:
- calculate savings growth
- calculate interest income
- track savings goals
- estimate how many months are needed to reach a goal
""")

    with gr.Row():

        with gr.Column():

            gr.Markdown("## Saving Goal")

            saving_goal_name = gr.Textbox(
                label="What Are You Saving For?",
                placeholder="Example: House"
            )

            saving_goal_amount = gr.Number(
                label="Saving Goal Amount"
            )

            years_to_goal = gr.Number(
                label="Years To Reach Goal"
            )

            monthly_goal_saving = gr.Number(
                label="How Much You Want To Save Monthly"
            )

        with gr.Column():

            gr.Markdown("## Income and Expenses")

            monthly_income = gr.Number(
                label="Monthly Income"
            )

            groceries = gr.Number(
                label="Groceries Expense"
            )

            transport = gr.Number(
                label="Transport Expense"
            )

            entertainment = gr.Number(
                label="Entertainment Expense"
            )

            rent_or_bills = gr.Number(
                label="Rent or Bills"
            )

            other_expenses = gr.Number(
                label="Other Monthly Expenses"
            )

        with gr.Column():

            gr.Markdown("## Interest and Results")

            bank_interest_rate = gr.Number(
                label="Bank Interest Rate (%)"
            )

            result_view = gr.Dropdown(
                choices=["Monthly", "Yearly"],
                label="Choose Result View",
                value="Yearly"
            )

    submit_button = gr.Button(
        "Calculate Savings Plan"
    )

    gr.Markdown("## Results")

    savings_advice = gr.Textbox(
        label="Savings Advice",
        lines=8
    )

    savings_table = gr.Dataframe(
        label="Savings Growth Table"
    )

    goal_time_text = gr.Textbox(
        label="Months Needed To Reach Goal",
        lines=5
    )

    final_result = gr.Textbox(
        label="Final Full Savings Plan",
        lines=18
    )

    submit_button.click(
        fn=savings_growth_planner,

        inputs=[
            saving_goal_name,
            saving_goal_amount,
            years_to_goal,
            monthly_goal_saving,
            monthly_income,
            groceries,
            transport,
            entertainment,
            rent_or_bills,
            other_expenses,
            bank_interest_rate,
            result_view
        ],

        outputs=[
            savings_advice,
            savings_table,
            goal_time_text,
            final_result
        ]
    )

app.launch()
```
